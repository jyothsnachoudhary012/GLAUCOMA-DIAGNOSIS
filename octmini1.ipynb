```python
#data augmentation
from tensorflow.keras.preprocessing.image import ImageDataGenerator
```

```python
train_gen=ImageDataGenerator(rescale=(1./255),horizontal_flip=True)
test_gen=ImageDataGenerator(rescale=(1./255))
```

```python
train=train_gen.flow_from_directory('/content/training',target_size=(120,120),class_mode='categorical',batch_size=8)
test=test_gen.flow_from_directory('/content/testing',target_size=(120,120),class_mode='categorical',batch_size=8)
```

```python
train.class_indices
```

```python
from tensorflow.keras.layers import Convolution2D, MaxPooling2D, GlobalAveragePooling2D, Dense
from tensorflow.keras.models import Sequential

model = Sequential()
model.add(Convolution2D(20, (3, 3), activation='relu', input_shape=(120, 120, 3)))
model.add(MaxPooling2D(pool_size=(2, 2)))
model.add(Convolution2D(40, (3, 3), activation='relu'))
model.add(MaxPooling2D(pool_size=(2, 2)))
model.add(Convolution2D(60, (3, 3), activation='relu'))
model.add(MaxPooling2D(pool_size=(2, 2)))
model.add(GlobalAveragePooling2D())
model.add(Dense(128, activation='relu'))
model.add(Dense(2, activation='sigmoid'))
```

```python
model.compile(optimizer='adam',loss='binary_crossentropy',metrics=['accuracy'])
```

```python
model.fit(train,batch_size=8,validation_data=test,epochs=10)
```

```python
def glaucoma_prediction(test_image):
  image = img_to_array(test_image)
  image = np.expand_dims(image, axis = 0)
  result = np.argmax(model.predict(image))
  return result
```

```python
def calculate_accuracy(true_labels, predicted_labels):
    if len(true_labels) != len(predicted_labels):
        raise ValueError("Lengths of true labels and predicted labels must be the same.")

    correct_count = 0
    total_count = len(true_labels)

    for true_label, predicted_label in zip(true_labels, predicted_labels):
        if true_label == predicted_label:
            correct_count += 1

    accuracy = float((correct_count / total_count)*100)


    return accuracy

# Example usage:
true_labels = ["Negative", "Positive", "Negative", "Positive", "Positive"]
predicted_labels = ["Negative", "Positive", "Negative", "Negative", "Positive"]
# Convert labels to binary representation (0 for negative, 1 for positive)
true_labels_binary = [0 if label == "Negative" else 1 for label in true_labels]
predicted_labels_binary = [0 if label == "Negative" else 1 for label in predicted_labels]
accuracy = calculate_accuracy(true_labels_binary, predicted_labels_binary)
print("Accuracy:", accuracy)
```
